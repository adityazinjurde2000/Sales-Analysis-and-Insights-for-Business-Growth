# 1. Data Cleaning and Preparation:
## Step 1.1: Handle Missing Values

```python
# Check for missing values in the dataset
print(df.isnull().sum())
```

```python
# For numerical columns (like 'Sales_Amount')
df['Sales_Amount'] = df['Sales_Amount'].fillna(df['Sales_Amount'].mean())
df['Quantity_Sold'] = df['Quantity_Sold'].fillna(df['Quantity_Sold'].mean())
df['Discount_Applied'] = df['Discount_Applied'].fillna(df['Discount_Applied'].mean())
df['Profit_Margin'] = df['Profit_Margin'].fillna(df['Profit_Margin'].mean())
```

```python
# For categorical columns (like 'Product_Category')
df['Product_Category'] = df['Product_Category'].fillna(df['Product_Category'].mode()[0])
```

```python
# Step 3: Check missing values again after filling them
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
df.info
```

 ## Step 1.2: Format the Date Column

```python
# Step 1: Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
# Step 2: Extract useful features
df['Year'] = df['Date'].dt.year      
df['Month'] = df['Date'].dt.month   
df['Day_of_Week'] = df['Date'].dt.dayofweek  
```

```python
# Display the first few rows to check the new features
print("\nData after date formatting and feature extraction:")
print(df.head())
```

```python
df.head()
```

 ## Step 3: Check for Outliers


```python
# Calculate Q1 and Q3 for the columns that are numerical
Q1 = df[['Sales_Amount', 'Quantity_Sold', 'Discount_Applied', 'Profit_Margin']].quantile(0.25)
Q3 = df[['Sales_Amount', 'Quantity_Sold', 'Discount_Applied', 'Profit_Margin']].quantile(0.75)
```

```python
# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Find outliers (values outside the range [Q1 - 1.5*IQR, Q3 + 1.5*IQR])
outliers = ((df[['Sales_Amount', 'Quantity_Sold', 'Discount_Applied', 'Profit_Margin']] < (Q1 - 1.5 * IQR)) | 
            (df[['Sales_Amount', 'Quantity_Sold', 'Discount_Applied', 'Profit_Margin']] > (Q3 + 1.5 * IQR)))

```

```python
# Count the number of outliers in each column
print("\nOutliers in numerical columns:")
print(outliers.sum())
```

```python
# Remove rows with outliers (optional, if you want to remove them)
df_no_outliers = df[~outliers.any(axis=1)]
```

```python
# Alternatively, cap outliers to the nearest value (optional)
for col in ['Sales_Amount', 'Quantity_Sold', 'Discount_Applied', 'Profit_Margin']:
    df[col] = df[col].clip(lower=Q1[col] - 1.5 * IQR[col], upper=Q3[col] + 1.5 * IQR[col])

```

```python
print("\nData after cleaning outliers:")
print(df_no_outliers.head()) 
```

# 4. Customer Insights:

 ## 4.1 Identify the age groups generating the most sales and profit.

```python
# Group by Customer_Age_Group to calculate total sales and profit
age_group_analysis = df.groupby('Customer_Age_Group').agg({
    'Sales_Amount': 'sum',
    'Profit_Margin': 'sum'
}).reset_index()

# Sort by total sales
age_group_analysis = age_group_analysis.sort_values(by='Sales_Amount', ascending=False)

# Display the results
print("Sales and Profit by Age Group:")
print(age_group_analysis)

# Plot sales and profit by age group
plt.figure(figsize=(10, 5))
x = age_group_analysis['Customer_Age_Group']
sales = age_group_analysis['Sales_Amount']
profits = age_group_analysis['Profit_Margin']

plt.bar(x, sales, label='Total Sales', alpha=0.7, color='blue')
plt.bar(x, profits, label='Total Profit', alpha=0.7, color='orange', bottom=sales-sales)  # Overlay profits
plt.title('Sales and Profit by Age Group')
plt.xlabel('Customer Age Group')
plt.ylabel('Amount')
plt.legend()
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

 ## 4.2Analyze how age groups respond to discounts.

```python
# Group by Customer_Age_Group and calculate average discount
age_group_discount = df.groupby('Customer_Age_Group').agg({
    'Discount_Applied': 'mean',
    'Sales_Amount': 'sum'
}).reset_index()

# Sort by sales to identify the age groups responding the most
age_group_discount = age_group_discount.sort_values(by='Sales_Amount', ascending=False)

# Display the results
print("Average Discount and Sales by Age Group:")
print(age_group_discount)

# Plot average discount vs sales for each age group
fig, ax1 = plt.subplots(figsize=(10, 5))

# Bar chart for sales
ax1.bar(age_group_discount['Customer_Age_Group'], age_group_discount['Sales_Amount'], alpha=0.7, color='skyblue', label='Sales')
ax1.set_xlabel('Customer Age Group')
ax1.set_ylabel('Sales Amount', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticks(range(len(age_group_discount['Customer_Age_Group'])))
ax1.set_xticklabels(age_group_discount['Customer_Age_Group'], rotation=45)

# Line plot for average discount
ax2 = ax1.twinx()
ax2.plot(age_group_discount['Customer_Age_Group'], age_group_discount['Discount_Applied'], color='red', marker='o', label='Average Discount')
ax2.set_ylabel('Average Discount (%)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Add legends
fig.legend(loc="upper right", bbox_to_anchor=(0.9, 0.9))
plt.title('Average Discount and Sales by Age Group')
plt.grid(axis='y')
plt.show()
```

 # 5. Visualization:

## 5.1 Sales trends over time.

```python
# Ensure the 'Date' column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Extract Year and Month for grouping
df['YearMonth'] = df['Date'].dt.to_period('M')

# Group data by Year-Month to calculate total sales
sales_trends = df.groupby('YearMonth')['Sales_Amount'].sum().reset_index()
sales_trends['YearMonth'] = sales_trends['YearMonth'].astype(str)  # Convert period to string

# Plot Sales Trends
plt.figure(figsize=(12, 6))
plt.plot(sales_trends['YearMonth'], sales_trends['Sales_Amount'], marker='o', color='blue')
plt.title('Sales Trends Over Time', fontsize=16)
plt.xlabel('Time (Year-Month)', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## 5.2 Top-performing products and regions.

```python
# Top-performing Products
top_products = df.groupby('Product_ID')['Sales_Amount'].sum().nlargest(5).reset_index()

# Plot Top Products
plt.figure(figsize=(10, 5))
plt.bar(top_products['Product_ID'], top_products['Sales_Amount'], color='green', alpha=0.8)
plt.title('Top 5 Performing Products', fontsize=16)
plt.xlabel('Product ID', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.grid(axis='y')
plt.show()

# Top-performing Regions
top_regions = df.groupby('Region')['Sales_Amount'].sum().sort_values(ascending=False).reset_index()

# Plot Top Regions
plt.figure(figsize=(10, 5))
plt.bar(top_regions['Region'], top_regions['Sales_Amount'], color='purple', alpha=0.8)
plt.title('Top Performing Regions', fontsize=16)
plt.xlabel('Region', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

 ## 5.3 Correlation between discounts and sales.

```python
# Scatter Plot: Discounts vs Sales
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Discount_Applied', y='Sales_Amount', alpha=0.6, color='orange')
plt.title('Correlation Between Discounts and Sales', fontsize=16)
plt.xlabel('Discount Applied (%)', fontsize=12)
plt.ylabel('Sales Amount', fontsize=12)
plt.grid(True)
plt.show()

# Calculate Correlation
correlation = df[['Discount_Applied', 'Sales_Amount']].corr().iloc[0, 1]
print(f"Correlation between Discount and Sales: {correlation:.2f}")
```
